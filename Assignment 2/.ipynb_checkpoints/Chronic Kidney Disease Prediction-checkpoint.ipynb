```python
import numpy as np
import pandas as pd
from sklearn import preprocessing
from sklearn.model_selection import train_test_split
import matplotlib.pyplot as plt
from sklearn.metrics import accuracy_score
import seaborn as sns
from sklearn.tree import DecisionTreeClassifier
from sklearn import svm
from sklearn.ensemble import RandomForestClassifier
```

```python
kidney = pd.read_csv('kidney_disease.csv')
```

### Data Preparation

```python
# Mapping the text to 1/0 and cleaning the dataset 
kidney[['htn','dm','cad','pe','ane']] = kidney[['htn','dm','cad','pe','ane']].replace(to_replace={'yes':1,'no':0})
kidney[['rbc','pc']] = kidney[['rbc','pc']].replace(to_replace={'abnormal':1,'normal':0})
kidney[['pcc','ba']] = kidney[['pcc','ba']].replace(to_replace={'present':1,'notpresent':0})
kidney[['appet']] = kidney[['appet']].replace(to_replace={'good':1,'poor':0,'no':np.nan})
kidney['classification'] = kidney['classification'].replace(to_replace={'ckd':1.0,'ckd\t':1.0,'notckd':0.0,'no':0.0})
kidney.rename(columns={'classification':'class'},inplace=True)

kidney['pe'] = kidney['pe'].replace(to_replace='good',value=0) # Not having pedal edema is good
kidney['appet'] = kidney['appet'].replace(to_replace='no',value=0)
kidney['cad'] = kidney['cad'].replace(to_replace='\tno',value=0)
kidney['dm'] = kidney['dm'].replace(to_replace={'\tno':0,'\tyes':1,' yes':1, '':np.nan})
```

```python
data = kidney
total = data.isnull().sum().sort_values(ascending=False)
percent = (data.isnull().sum()/data.isnull().count()).sort_values(ascending=False)
missing_data = pd.concat([total, percent], axis=1, keys=['Total', 'Percent'])
plt.subplots(figsize=(15, 6))
sns.barplot(x=missing_data.index, y=missing_data['Percent'])
plt.xlabel('Features', fontsize=15)
plt.ylabel('Percent of missing values', fontsize=15)
plt.title('Percent missing data by feature', fontsize=15)
missing_data.head()
```

```python
kidney = data.dropna()
```

```python
kidney['pcv'] = kidney['pcv'].astype('int64')
kidney['wc'] = kidney['wc'].astype('int64')
kidney['rc'] = kidney['rc'].astype('float64')
kidney['dm'] = kidney['dm'].astype('int64')
kidney['class'] = kidney['class'].astype('int64')
```

```python
kidney = kidney.drop(['id'],axis=1)
kidney = kidney.reset_index()
```

### Now applying different classifiers on the dataset

- Naive Bayes
- Decision Tree
- Random Forest
- SVM

#### 1. Naive Bayes

```python
def prior_prob(y_train,label):
    
    """This function calculates the prior probability of a label/class i.e. P(Y = label)"""
    
    total_examples = y_train.shape[0]
    class_examples = np.sum(y_train == label)
    
    return (class_examples)/float(total_examples)


def cond_prob(x_train,y_train,feature_col,feature_val,label):
    
    x_filtered = x_train[y_train == label]
    numerator = np.sum(x_filtered[:,feature_col] == feature_val)
    denominator = np.sum(y_train == label)
    
    return (numerator+1)/float(denominator)
```

```python
def predict_naive_bayes(x_train,y_train,x_test):
    
    classes = np.unique(y_train)
    n_features = x_train.shape[1]
    
    post_probs = []
    
    for label in classes:
        likelihood = 1.0
        
        for f in range(n_features):
            cond = cond_prob(x_train,y_train,f,x_test[f],label)

            likelihood *= cond
            
        prior = prior_prob(y_train,label)
        post = likelihood * prior
        
        post_probs.append(post)
    
    pred = np.argmax(post_probs)
    
    return pred
```

```python
def pred_list(X_train,X_test,Y_train):
    
    m = X_test.shape[0]
    
    y_pred = []
    
    for i in range(m):
        pred = predict_naive_bayes(X_train,Y_train,X_test[i])
        y_pred.append(pred)
    

    return y_pred
```

```python
y_pred_nb = pred_list(x_train,x_test,y_train)
```

#### 2. Decision Tree

```python
sk_tree = DecisionTreeClassifier(criterion='entropy',max_depth=5)
```

```python
sk_tree.fit(x_train,y_train)
```

```python
y_pred_dt = sk_tree.predict(x_test)
```

#### 3. Random Forest

```python
rf = RandomForestClassifier(n_estimators = 3, max_depth=5,criterion='entropy')
```

```python
rf.fit(x_train,y_train)
```

```python
y_pred_rf = rf.predict(x_test)
```

#### 4. SVM

```python
sv = svm.SVC()
```

```python
sv.fit(x_train,y_train)
```

```python
y_pred_sv = sv.predict(x_test)
```

### Checking accuracy

```python
acc_nb = accuracy_score(y_test,y_pred_nb)
acc_dt = accuracy_score(y_test,y_pred_dt)
acc_rf = accuracy_score(y_test,y_pred_rf)
acc_sv = accuracy_score(y_test,y_pred_sv)
```

```python
print("Accuracy using Naive Bayes - ",acc_nb)
print("Accuracy using Decision tree - ",acc_dt)
print("Accuracy using Random Forest - ",acc_rf)
print("Accuracy using SVM - ",acc_sv)
```
